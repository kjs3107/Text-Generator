```python
#for converting text to numbers we will import tokeniser
```

```python
#importing tensorflow libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer() #make a tokeniser object
```

```python
tokenizer.fit_on_texts([faqs])#to save our model in tokeniser
```

```python
tokenizer.word_index #for checking the no. given to ur text
```

```python
len(tokenizer.word_index)#checking our article length
```

```python
for sentence in faqs.split('\n'):
  print(sentence)
```

```python
for sentence in faqs.split('\n'):
  print(tokenizer.texts_to_sequences([sentence])[0]) # as double list is created so to avoid that we used[0]
```

```python
#we will check out as it is correct or not taking our first one i.e.[[29,11,10,5]] and then checking these values in tokeniser so the word coming out from this is what decision making is
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
[len(x) for x in input_sequences]
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
#zero padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')#storing in a variable
```

```python
padded_input_sequences
```

```python
padded_input_sequences[:,:-1]#for input
```

```python
X = padded_input_sequences[:,:-1]
```

```python
padded_input_sequences[:,-1]#for output
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
#as a classification task:-Binary or multiclass we use multiclass classification and in that we will do one hot encoding for transforming
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=256)
```

```python
y.shape
```

```python
y
```

```python
#embedding layer:- dence vector for every word for better representation after that lstm and after that for output as for softmax we will use dense layer
```

```python
#importing necessary libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(256, 100, input_length=76))
model.add(LSTM(150))
model.add(Dense(256, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
X
```

```python
model.fit(X,y,epochs=100)
```

```python
text = "Improve"
#Tokenisation
token_text = tokenizer.texts_to_sequences([text])[0]
#padding
padded_token_text = pad_sequences([token_text],maxlen=76,padding='pre')
print(padded_token_text)
```

```python
text = "Improve"
#Tokenisation
token_text = tokenizer.texts_to_sequences([text])[0]
#padding
padded_token_text = pad_sequences([token_text],maxlen=76,padding='pre')
#predict
model.predict(padded_token_text)
```

```python
model.predict(padded_token_text).shape
```
